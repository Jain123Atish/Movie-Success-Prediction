Categorising the Movie rating into 3 classes Hit,Avg,Flop
Here I have dataset named movie_metadata in which the target variable is Movie_score and other variables that decide the Movie_score. Instead of just Movie_score,With the help of other parameters I want to predict whether a Movie is Hit,Avg or Flop.



|Movie_score | Classify |
| --- | ---|
|1-3 | Flop Movie|
|3-6 | Average Movie|
|6-10 | Hit Movie|

# Importing necessary Libraries

```python
#Importing necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

# Load Data

```python
data=pd.read_csv('movie_metadata.csv')
data.head()
```

# Categorizing the target varible 

Here we are categorizing the target variable in such a way that Movie_score between 1 and 3 is FLOP , between 3 and 6 is AVG, between 6 and 10 is HIT.


```python
#Categorising the target varible 
bins = [ 1, 3, 6, 10]
labels = ['FLOP', 'AVG', 'HIT']
data['Movie_Success'] = pd.cut(data['Movie_score'], bins=bins, labels=labels)
```

```python
data.groupby(['Movie_Success']).size().plot(kind="bar",fontsize=14)
plt.xlabel('Categories')
plt.ylabel('Number of Movies')
plt.title('Categorization of Movies')
```

We can see a new column named Movie_Success correctly categorising the Movie_score

# Label Encoding

All the categorical columns and the columns with text data are being Label Encodeded in this step.

```python
#Label encoding the categorical columns
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
list=['color', 'director_name', 'actor_2_name',
        'genres', 'actor_1_name',
        'actor_3_name',
        'plot_keywords',
        'language', 'country', 'content_rating',
       'title_year', 'aspect_ratio','Movie_Success','movie_title']
data[list]=data[list].apply(lambda x:le.fit_transform(x))

#A sample of data after label encoding
data.head()
```

# Classification Model Building

Splitting the data into x and y where x contains Indepentent variables and y contain Target/Dependent variable.


```python
x=data.drop(columns=['Movie_Success','color','duration','country'],axis=1) #independent Variable 
y=data['Movie_Success'] #dependent / target varible

print('Shape of x=',x.shape)
print('Shape of y=',y.shape)
```

# Split Data

```python
#Spliting the data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.8, random_state = 551)
print('Shape of X_train = ',x_train.shape)
print('Shape of y_train = ',y_train.shape)
print('Shape of X_test = ',x_test.shape)
print('Shape of y_test = ',y_test.shape)
```

```python
data.head()
```

# Features Selection

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x_train)
x_train = sc.transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```
